```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
crashes = sns.load_dataset("car_crashes")
```

```python
crashes
```

```python
def bar_plot():
    plt.figure(figsize=(10, 6))
    sns.barplot(x="abbrev", y="total", data=crashes)
    plt.title("Total Number of Car Crashes by State")
    plt.xlabel("State Abbreviation")
    plt.ylabel("Total Crashes")
    plt.show()
bar_plot()
```

```python
def scatter_plot():
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x="alcohol", y="total", data=crashes)
    plt.title("Relationship Between Alcohol-Related Crashes and Total Crashes")
    plt.xlabel("Alcohol-Related Crashes")
    plt.ylabel("Total Crashes")
    plt.show()
scatter_plot()
```

```python
def histogram():
    plt.figure(figsize=(10, 6))
    sns.histplot(crashes["total"], bins=15, kde=True)
    plt.title("Distribution of Total Crashes")
    plt.xlabel("Total Crashes")
    plt.ylabel("Frequency")
    plt.show()
histogram()
```
